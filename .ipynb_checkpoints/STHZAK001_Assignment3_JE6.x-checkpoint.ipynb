# DSB-SC AM (Modulation and Demodulation)

# Exercise 6.1a - DSB-SC AM (Modulation)

```julia
using Pkg
using PyPlot
using FFTW
```

```julia
fs = 100e3
fm = 1e3
fc = 20e3

Δt = 1 / fs
tm = 3 / fm
t = 0: Δt: tm

f(t) = cos.(2π * fm.* t);
xm = f.(t);

plot(t, xm);
title("Modulation Waveform f(t)");
ylabel("f(t)")
xlabel("t")
```

```julia
FM = fft(xm)

N = length(t)
Δf = fs / N
fx = (-fs / 2 + Δf): Δf: (fs / 2);

plot(fx, fftshift(abs.(FM)));
xlabel("fft(xm)");
xlabel("Frequency In Hz");
```

```julia
xc = cos.(2π * fc.* t)

plot(t, xc);
title("Carrier Sinusoid (cos(2πfct))");
ylabel("xc");
xlabel("t");
```

```julia
XC = fft(xc);

plot(fx, fftshift(abs.(XC)));
ylabel("XC");
xlabel("Frequency In Hz")
```

```julia
ϕ = xm.* xc;

plot(t, ϕ);
title("Modulated Carrier Wave");
ylabel("ϕ(t)");
xlabel("t");
```

```julia
Φ = fft(ϕ)

plot(fx, abs.(Φ))
title("Modulated Carrier Wave")
ylabel("Φ")
xlabel("Frequency In Hz");
```

```julia
square_wave(t) = ((sin.(2π * fm *t ) .> 0) .-0.5) * 2;
f_square = square_wave(t);

plot(t, f_square);
ylabel("f_sqaure");
xlabel("t");
```

```julia
F_Square = fft(f_square);
plot(fx, fftshift(abs.(F_Square)));
```

```julia
ϕ_square = f_square.* xc;

plot(t, ϕ_square);
ylabel("ϕ_sqaure");
xlabel("t");
```

```julia
Φ_square = fft(ϕ_square);

plot(fx, fftshift(abs.(Φ_square)));
ylabel("Φ_square");
xlabel("fx");
```

# Exercise 6.1b - DSB-SC Demodulation

```julia
y = ϕ.* xc;

plot(t, y);
ylabel("Y");
xlabel("t");
```

```julia
Y = fft(y);

plot(fx, fftshift(abs.(Y)));
ylabel("Y");
xlabel("fx");
```

```julia
rect(t) = (abs.(t).<= 0.5) * 1.0;
H = rect(fx ./ 3000);
V = fftshift(H) .* Y;

plot(fx, fftshift(abs.(V)));
ylabel("V");
xlabel("Frequency In Hz");
```

```julia
v = real.(ifft(V));

plot(t, v);
ylabel("V");
xlabel("t");
```

# Exercise 6.1c - Effect of Phase Error In Demodulation of DSC-SC

```julia
y_30 = ϕ .* cos.(2π * fc.* t .+ π / 6);
Y_30 = fft(y_30);

V_30 = fftshift(H) .* Y_30;
v_30 = real.(ifft(V_30));

plot(t, v_30);
ylabel("Phase Error of 30 Degrees");
xlabel("t");
```

# Exercise 6.1d - DSB-SC AM With Noise

# Exercise 6.3 - Quadrature Multiplexing 

```julia
fm1 = 200;
fm2 = 1000;
fc2 = 10000;

x1 = cos.(2π * fm1 * t);
x2 = cos.(2π * fm2 * t);

ϕ1 = x1 .* cos.(2π * fc2 * t);
ϕ2 = x2 .* sin.(2π * fc2 * t);

ϕm = ϕ1 .+ ϕ2;

y1 = ϕm .* cos.(2π * fc2 * t);
y2 = ϕm .* sin.(2π * fc2 * t);

Y1 = fft(y1);
Y2 = fft(y2);

Hm = rect(fx ./ 20000)

E1 = fftshift(Hm) .* Y1
E2 = fftshift(Hm) .* Y2

e1 = real.(ifft(E1));
e2 = real.(ifft(E2));

plot(t, x1);
plot(t,  ϕ1);
xlabel("t")
```

```julia
plot(t, y1);
plot(t, e1);
xlabel("t")
```

```julia
plot(t, x1);
plot(t,  ϕ1);
plot(t, y1);
plot(t, e1);
xlabel("t")
```

```julia
plot(t, x2);
plot(t,  ϕ2);
xlabel("t")
```

```julia
plot(t, y2);
plot(t, e2);
xlabel("t")
```

```julia
plot(t, x2);
plot(t,  ϕ2);
plot(t, y2);
plot(t, e2);
xlabel("t")
```
